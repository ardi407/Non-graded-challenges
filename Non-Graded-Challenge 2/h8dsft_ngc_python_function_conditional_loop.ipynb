```python
def number_of_word(sentence):
    words = sentence.split(' ')
    return (len(words))
    
```

```python
def find_multiple(start_range, end_range, multiple):
    result = []
    for number in range(start_range, end_range):
        if (number * multiple) <= end_range:
            calc = number * multiple
            result.append(calc)
    return result
```

```python
find_multiple(1,50,10)
```

```python
def Grade_Converter(convert_to, **kwargs):
    '''
    This function takes the parameter:
    1. convert_to such as 'letter_grade' or 'gpa depending on what you need
    2. kwargs that have key as string: "name" and value as an integer score.
    '''
    result = {}
    if convert_to == 'letter_grade':
        for name, grade in kwargs.items():
            if grade >=85:
                result[name]='A'
            elif grade >=70:
                result[name] ='B'
            elif grade >=60:
                result[name]='C'
            elif grade >=50:
                result[name]='D'
            elif grade >=40:
                result[name]='E'
            else:
                result[name]='F'
    elif convert_to == 'gpa':
        for name, grade in kwargs.items():
            if grade >=85:
                result[name]=4
            elif grade >=70:
                result[name] =3
            elif grade >=60:
                result[name]=2.5
            elif grade >=50:
                result[name]=2
            elif grade >=40:
                result[name]=1.5
            else:
                result[name]=1
    else:
        print("Perintah salah")
    return result
```

```python
Grade_Converter(convert_to='letter_grade', Albert=90, Dwi=82, Syahdan=58, Veronica=84)
```

```python
def get_fibonacci(n):
    '''
    this function will generate a fibonacci number based on the sequence(n)
    this function only takes one number indicating the sequence number
    '''
    fibonacci = [0,1,1]
    list_n=[]
    for i in range(0,n):
        list_n.append(i)
        next_number = fibonacci[-1]+fibonacci[-2]
        fibonacci.append(next_number)
    return fibonacci[len(list_n)]
```

```python
get_fibonacci(10)
```
