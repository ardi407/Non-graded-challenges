```python
from google.colab import auth
from google.cloud import bigquery
import pandas as pd

auth.authenticate_user()
project_id = "ngc-6-400305" # <-- sesuaikan dengan project_id kalian
client = bigquery.Client(project=project_id)

# put your query here
query = """
SELECT *
FROM bigquery-public-data.austin_bikeshare.bikeshare_trips
"""
      # Project ID             #dataset       #table
# Set up the query
query_job = client.query(query).to_dataframe()
query_job
```

```python
query = """
SELECT table_name, COUNT(table_name) AS jumlah
FROM `bigquery-public-data.austin_bikeshare.INFORMATION_SCHEMA.TABLES`
GROUP BY table_name
"""

# Set up the query
query_job = client.query(query).to_dataframe()
query_job
```

Terdapat 2 tables dalam austin_bikeshare dataset, yaitu:

```python
query = """
SELECT COUNT(data_type) as string_data_type
FROM `bigquery-public-data.austin_bikeshare.INFORMATION_SCHEMA.COLUMNS`
WHERE table_name = 'bikeshare_stations' AND data_type = 'STRING'
"""
# WHERE table_name = 'bikeshare_stations';
# Set up the query
query_job = client.query(query).to_dataframe()
query_job
```

**Terdapat 7 Column** yang memiliki datatype STRING.

```python
query = """
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE bike_id IS NOT NULL AND duration_minutes > 1000 AND
      subscriber_type = 'Walk Up' AND end_station_name = 'Stolen';
"""

# Set up the query
query_job = client.query(query).to_dataframe()
query_job
```
